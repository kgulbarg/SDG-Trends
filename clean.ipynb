```python
def clean_melt_df(xlFile):
    df = pd.read_excel(xlFile)
    df.info()
    df.drop(columns=["Goal", "Target", "Indicator", "SeriesCode", "SeriesDescription","Observation Status", "Reporting Type"], inplace=True)
    df.info()
    # transforming wide to long table
    year_cols = [str(y) for y in range(2015, 2025)]
    fixed_columns = ["GeoAreaCode", "GeoAreaName", "Age", "Location", "Sex", "Units"]

    df_long = df.melt(
        id_vars=fixed_columns,
        value_vars=year_cols,
        var_name="Year",
        value_name="Value"
    )

    df_long.head()
    return df_long
```
